```python
#Total Charges is in object because there afre blank values in it ---fill blank with 0 and change the dtype to float
```

```python
#Senior citizens values are in o and 1 convert it into no or yes
```

```python
def conv(values):
    if values==1:
        return "Yes"
    else:
        return "No"
    
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
        
```

```python
df["SeniorCitizen"]
df
```
